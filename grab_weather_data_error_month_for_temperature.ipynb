```python
import iris
import datetime as dt
import time
import numpy as np
import bz2
import json
import wget
import os
import bz2
import urllib

ll_lat = 365
ur_lat = 510
ll_lon = 375
ur_lon = 485

raw_out = './raw_data/grib/'

base_path = 'ftp://ftp-cdc.dwd.de/pub/REA/COSMO_REA6/hourly/2D/'
var = 'TOT_PRECIP'
month = 8
arr = []

if month >= 10:
    month_string = str(month)
else:
    month_string = '0' + str(month)

time_index = open('./processed_data/weather/{}.2D.2015{}.txt'.format(var, month_string), 'w')
path = base_path + var + '/' + '{}.2D.2015{}.grb.bz2'.format(var, month_string)

print(path)

_ = wget.download(path, out=raw_out)

filename = raw_out + '{}.2D.2015{}.grb.bz2'.format(var, month_string)

print('download finished')

zipfile = bz2.BZ2File(filename) # open the file
data = zipfile.read() # get the decompressed data
newfilepath = filename[:-4] # new file name

with open(newfilepath, 'wb') as f:
    f.write(data)

print('unzip finished')

```

```python
cubes = iris.load(newfilepath)
```

```python
arr = []
for i in range(cubes[1].shape[0]):
    print('cube {}'.format(i))
    timestamp = time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime(cubes[1][i].coords('time')[0].cell(0).point.round()*3600))
    data = cubes[1][i].data[ll_lat:ur_lat, ll_lon:ur_lon]
#             print(data)
    arr.append(data)
    time_index.write(timestamp + '\n')

print('data extracted')

del(cubes)

time_index.close()
arr = np.array(arr)
np.save('./processed_data/weather/{}.2D.2015{}.npy'.format(var, month_string), arr)

os.remove(filename)
os.remove(newfilepath)
```
